```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = '../DataSets/2022-2023FootballPlayerStats.csv'
df = pd.read_csv(ruta_archivo, encoding='ISO-8859-1', delimiter=';')

# Ver las primeras filas del dataset
print(df.head())
```

```python
#Grafico de lineas -Escala de tiempo en x
#Grafico scatter plot- Analisis de dos variables - grafico bivariado
#Grafico de barras - Se utiliza para variables categoricas
#Grafico Histograma - Se utiliza para variables numericas continuas, ver frecuencia
#Grafico Boxplot - Se utiliza para analisis de outliers
#Grafico Piechart - Se utiliza para variables categoricas
```

```python
# Grafico de lineas - Escala de tiempo en x

fig, ax = plt.subplots(figsize=(12, 4))

# Agrupar los datos por squad y sumar los goles
goals_by_squad = df.groupby('Squad')['Goals'].sum().reset_index()

# Ordenar el dataframe por los goles en orden descendente
goals_by_squad_sorted = goals_by_squad.sort_values(by='Goals', ascending=False)

# Obtener los 5 equipos con más goles
top_5_squads = goals_by_squad_sorted.head(5)

# Graficar solo los 5 equipos con más goles
ax.bar(top_5_squads['Squad'], top_5_squads['Goals'], label='Goles Acumulados')

ax.set_xlabel('Squad')
ax.set_ylabel('Goles Acumulados')
ax.set_title('Top 5 Squads con mayores goles')
ax.legend()

plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad

plt.show()
```

```python
# Grafico scatter plot - Analisis de dos variables - grafico bivariado
plt.scatter(df['Goals'], df['Assists'])
plt.xlabel('Goles')
plt.ylabel('Asistencias')
plt.title('Relación entre goles y asistencias')
plt.show()
```

```python
# Grafico de barras - Se utiliza para variables categoricas
sns.countplot(x='Pos', data=df)
plt.xlabel('Posición')
plt.ylabel('Cantidad de jugadores')
plt.title('Distribución de jugadores por posición')
plt.show()
```

```python
# Grafico Histograma - Se utiliza para variables numericas continuas, ver frecuencia
plt.hist(df['PasTotCmp%'], bins=10)
plt.xlabel('Porcentaje de pases completados')
plt.ylabel('Frecuencia')
plt.title('Distribución del porcentaje de pases completados')
plt.show()
```

```python
# Grafico Boxplot - Se utiliza para analisis de outliers

df = pd.read_csv(ruta_archivo, encoding='ISO-8859-1', delimiter=';')

# Agrupar los datos por squad y sumar los goles
goals_by_squad = df.groupby('Squad')['Goals'].sum().reset_index()

# Ordenar el dataframe por los goles en orden descendente
goals_by_squad_sorted = goals_by_squad.sort_values(by='Goals', ascending=False)

# Obtener los 5 equipos con más goles
top_5_squads = goals_by_squad_sorted.head(5)

# Filtrar los datos para los 5 equipos seleccionados
filtered_data = df[df['Squad'].isin(top_5_squads['Squad'])]

# Crear el gráfico de boxplot
fig, ax = plt.subplots(figsize=(12, 4))
ax.boxplot(filtered_data.groupby('Squad')['Goals'].apply(list))

# Configurar el gráfico
plt.title("Cantidad de goles por equipo")
plt.xlabel("Equipo")
plt.ylabel("Cantidad de goles")

# Etiquetas de los equipos en el eje x
ax.set_xticklabels(filtered_data['Squad'].unique())

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por squad y sumar los GCA
gca_by_squad = df.groupby('Squad')['GCA'].sum().reset_index()

# Ordenar el dataframe por los GCA en orden descendente
gca_by_squad_sorted = gca_by_squad.sort_values(by='GCA', ascending=False)

# Obtener los 5 equipos con más GCA
top_5_squads = gca_by_squad_sorted.head(5)

# Filtrar los datos para los 5 equipos seleccionados
filtered_data = df[df['Squad'].isin(top_5_squads['Squad'])]

# Crear el gráfico de boxplot
fig, ax = plt.subplots(figsize=(12, 4))
ax.boxplot(filtered_data.groupby('Squad')['GCA'].apply(list))

# Configurar el gráfico
plt.title("Goal-creating actions por equipo")
plt.xlabel("Equipo")
plt.ylabel("Goal-creating actions (GCA)")

# Etiquetas de los equipos en el eje x
ax.set_xticklabels(filtered_data['Squad'].unique())

# Mostrar el gráfico
plt.show()
```
