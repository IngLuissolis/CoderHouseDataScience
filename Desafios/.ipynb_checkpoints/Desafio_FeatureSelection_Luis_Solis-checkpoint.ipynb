# Algoritmo de agrupamiento K-means

```python
import numpy as np
from sklearn.cluster import KMeans

from sklearn.metrics import davies_bouldin_score,silhouette_score,silhouette_samples,calinski_harabasz_score

from sklearn.preprocessing import MinMaxScaler, StandardScaler

import matplotlib.pyplot as plt
import matplotlib.cm as cm

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# Drop the columns that are not relevant for clustering
df = df[columns]
df.head()
```

```python
# Calcular estadísticas descriptivas
df.describe()
```

```python
# Handle missing values
df_filtered = df.fillna(0)

# normalizar data
X = df_filtered.values
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)

X
```

```python
# diccionarios vacios para ser almacenados los resultados
sse,db,slc,cal = {}, {}, {}, {}
for k in range(2, 10):
    # semilla de 10 para asegurar reproducibilidad
    kmeans = KMeans(n_clusters=k, max_iter=1000,random_state=10).fit(X) # Algoritmo K means
    if k == 3: labels = kmeans.labels_
    clusters = kmeans.labels_ # Extraccion de labels
    sse[k] = kmeans.inertia_ # Inercia: Suma de las distancias de las muestras a su centroide mas cercano
    db[k] = davies_bouldin_score(X,clusters) # Calculo del indice de Davies Bouldin
    slc[k] = silhouette_score(X,clusters) # Calculo del indice de Silhouette
    cal[k] = calinski_harabasz_score(X,clusters) # Calculo del indice de Calinski-Harabasz
```

## Metodo del Codo

```python
plt.figure(figsize=(15,10))
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Numero de clusters")
plt.ylabel("SSE")
plt.show()
```

### Davies-Bouldin Index

```python
plt.figure(figsize=(15,10))
plt.plot(list(db.keys()), list(db.values()))
plt.xlabel("Numero de clusters")
plt.ylabel("Indice Davies-Bouldin")
plt.show()
```

### Indice de Silhouette

```python
plt.figure(figsize=(15,10))
plt.plot(list(slc.keys()), list(slc.values()))
plt.xlabel("Numero de clusters")
plt.ylabel("Score de Silhouette")
plt.show()
```

```python
for n_clusters in range(2, 10):
    # Crear un subplot de 1 fila y 2 columnas
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    # El 1 subplot es para el score de silhouette 
    # El rango del Ind. silhouette es [-1, 1]
    ax1.set_xlim([-0.1, 1]) # para nuestro caso lo voy a dejar en este rango
    # El (n_clusters+1)*10 es para insertar un espacio en blanco entre la silueta
     # y el grafico de la nube de puntos
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])
    # Iniciamos el proceso de ajuste a los datos
    clusterer = KMeans(n_clusters=n_clusters,max_iter=1000, random_state=10)
    cluster_labels = clusterer.fit_predict(X)
    # El score de silhouette_score da el promedio de todos los valores 
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("Para n_clusters =", n_clusters,
          "El promedio del score de silhouette es: ", silhouette_avg)
    # Calcular el score de silhouette para cada muestra 
    sample_silhouette_values = silhouette_samples(X, cluster_labels)
    y_lower = 10
    for i in range(n_clusters):
        # Agregar los scores de silhouette para las muestras pertenecientes al cluster i y ordenarlos 
        ith_cluster_silhouette_values =sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort() # Ordenar
        size_cluster_i = ith_cluster_silhouette_values.shape[0] # Tamaño de clusters
        y_upper = y_lower + size_cluster_i 
        color = cm.nipy_spectral(float(i) / n_clusters) # Definir paleta de colores
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)
        # Etiquetar las graficas de silhouette con sus miembros de cluster en el medio
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        # Calcular el nuevo y_lower para la proxima grafica
        y_lower = y_upper + 10 
    ax1.set_title("Grafico de silhouette para varios clusters.")
    ax1.set_xlabel("Coeficientes de silhouette")
    ax1.set_ylabel("Label del Cluster")
    # La linea vertical para el promedio del score de silhouette de todos los valores
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])  # Quitar labels en y 
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1]) # definir ticks para comparacion

    # 2 grafica mostrando los clusters formados 
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters) # definir colores
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,c=colors, edgecolor='k') # scatterplot
    # Etiquetar a los clusters
    centers = clusterer.cluster_centers_
    # Dibujar en color blanco los centroides de los clusters
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',c="white", alpha=1, s=200, edgecolor='k')
    # Enumerar los centroides
    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,s=50, edgecolor='k')

    ax2.set_title("Visualizacion luego de obtener clusters")
    ax2.set_xlabel("Variable X")
    ax2.set_ylabel("Variable Y")

    plt.suptitle(("Analisis de Silhouette para algoritmo de KMeans clustering "
                  "con n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from numpy import unique
from numpy import where

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn import metrics
from sklearn.metrics.cluster import pair_confusion_matrix

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
```

```python
df = pd.read_csv(ruta_archivo, delimiter=',')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
# Drop the columns that are not relevant for clustering
df = df[columns]
df.head()
```

```python
df.info()
```

```python
#Perform PCA to obtain first 4 components for easy visualization
pca = PCA(4)
pca_data=pca.fit_transform(df)

#Perform Scaling
scaler = StandardScaler()
norm_data = scaler.fit_transform(pca_data)

norm_data = pd.DataFrame(data=norm_data, columns = columns)
norm_data.describe().T
```

## K-Means

```python
kmeans = KMeans(n_clusters=3, max_iter=500)
kmeans_labels = kmeans.fit_predict(norm_data)
```

```python
#Getting unique labels
centroids = kmeans.cluster_centers_
u_labels = np.unique(kmeans_labels)
 
#plotting the results:
 
for i in u_labels:
    filtered_data = norm_data[kmeans_labels==i]
    plt.scatter(filtered_data.iloc[:, 0] , filtered_data.iloc[:,1])
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, color = 'k')
plt.legend()
plt.show()
```

## DBSCAN

```python
# define the model
dbscan_model = DBSCAN(eps=0.25, min_samples=9)

# train the model
dbscan_model.fit(norm_data)

# assign each data point to a cluster
dbscan_labels = dbscan_model.labels_

# get all of the unique clusters
dbscan_clusters = unique(dbscan_labels)

for i in dbscan_clusters:
    # get data points that fall in this cluster
    filtered_data = norm_data[dbscan_labels==i]
    plt.scatter(filtered_data.iloc[:, 0] , filtered_data.iloc[:,1])
# show the DBSCAN plot
plt.legend()
plt.show()
```

## Gaussian Mixture Model

```python
# define the model
gaussian_model = GaussianMixture(n_components=3)

# train the model
gaussian_model.fit(norm_data)

# assign each data point to a cluster
gaussian_labels = gaussian_model.predict(norm_data)

# get all of the unique clusters
gaussian_clusters = unique(gaussian_labels)

for i in gaussian_clusters:
    # get data points that fall in this cluster
    filtered_data = norm_data[gaussian_labels==i]
    plt.scatter(filtered_data.iloc[:, 0] , filtered_data.iloc[:,1])
# show the plot
plt.legend()
plt.show()
```

## Evaluation metrics for clustering

### Silhouette Coefficient

```python
kmeans_silhouette = metrics.silhouette_score(norm_data, kmeans_labels, metric='euclidean')
dbscan_silhouette = metrics.silhouette_score(norm_data, dbscan_labels, metric='euclidean')
gaussian_silhouette = metrics.silhouette_score(norm_data, gaussian_labels, metric='euclidean')
```

### Calinski Harabasz Score

```python
kmeans_ch = metrics.calinski_harabasz_score(norm_data, kmeans_labels)
dbscan_ch = metrics.calinski_harabasz_score(norm_data, dbscan_labels)
gaussian_ch = metrics.calinski_harabasz_score(norm_data, gaussian_labels)
```

### Davies Bouldin Index

```python
kmeans_db = metrics.davies_bouldin_score(norm_data, kmeans_labels )
dbscan_db = metrics.davies_bouldin_score(norm_data, dbscan_labels)
gaussian_db = metrics.davies_bouldin_score(norm_data, gaussian_labels)
```

##  Comparision of clustering models

```python
data = [
    ['KMeans',kmeans_silhouette, kmeans_ch, kmeans_db], 
    ['DBSCAN', dbscan_silhouette, dbscan_ch, dbscan_db],
    ['Gaussian Mixture Model', gaussian_silhouette, gaussian_ch, gaussian_db]
       ]

result = pd.DataFrame(data, columns=['Model', 'Silhouette_Score', 'Calinski_Harabasz_Score', 'Davies_Bouldin_Index'])

result
```

### Basándome en los resultados de las métricas de clustering, las conclusiones son las siguientes:

    El modelo KMeans es el que mejor rendimiento tiene en general, con un Silhouette Score de 0.490543, un Calinski-Harabasz Score de 760.714819 y un Davies-Bouldin Index de 1.315154. Esto indica que los datos están bien agrupados en clusters, con una separación clara entre ellos.
    
    El modelo DBSCAN es el que peor rendimiento tiene, con un Silhouette Score de 0.008785, un Calinski-Harabasz Score de 73.881538 y un Davies-Bouldin Index de 2.286783. Esto indica que los datos no están bien agrupados en clusters, con una separación poco clara entre ellos.
    
    El modelo Gaussian Mixture Model tiene un rendimiento intermedio, con un Silhouette Score de 0.520297, un Calinski-Harabasz Score de 392.416142 y un Davies-Bouldin Index de 2.197281. Esto indica que los datos están bien agrupados en clusters, pero con una separación no tan clara como en el modelo KMeans.
    
    En conclusión, el modelo KMeans es el mejor modelo para el rendimiento de jugadores de fútbol. Este modelo es capaz de agrupar los datos en clusters de manera eficiente y clara, lo que permite identificar patrones y tendencias relevantes.

## Modelo K-Means (Opcion Alternativa)

```python
df = pd.read_csv(ruta_archivo, delimiter=',')
df.head()
```

```python
# Drop the columns that are not relevant for clustering
df_filtered = df[columns]

# Handle missing values
df_filtered = df_filtered.fillna(0)

# Scale the data
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_filtered)

# Select the number of clusters
n_clusters = 3

# Create the k-means model
kmeans = KMeans(n_clusters=n_clusters)

# Fit the model to the data
kmeans.fit(df_scaled)

# Predict the clusters for each data point
labels = kmeans.predict(df_scaled)

# Print the cluster labels
print(labels)
```

```python
# Create a scatter plot
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels)

# Add labels to the clusters
for i in range(n_clusters):
    plt.annotate(f"Cluster {i}", (df_scaled[labels == i, 0].mean(), df_scaled[labels == i, 1].mean()))
### Métricas de calidad para Clustering
# Show the plot
plt.show()
```

### Métricas de calidad para Clustering

```python
# Calculate the silhouette score
silhouette_score = silhouette_score(df_scaled, labels)

# Print the silhouette score
print(silhouette_score)### Métricas de calidad para Clustering
```

```python
# Calculate the Calinski-Harabasz score
ch_score = calinski_harabasz_score(df_scaled, labels)

# Print the Calinski-Harabasz score
print(ch_score)
```

```python
# Calculate the Davies-Bouldin score
db_score = davies_bouldin_score(df_scaled, labels)

# Print the Davies-Bouldin score
print(db_score)
```

### Resultados

    Basándome en los resultados de las métricas de clustering, las conclusiones son las siguientes:

    El modelo KMeans tiene un rendimiento muy bueno, con un Silhouette Score de 0.5584606596081799, un Calinski-Harabasz Score de 2575.912114796276 y un Davies-Bouldin Index de 1.0366205839889322.
    
    Estos resultados indican que los datos están bien agrupados en clusters, con una separación clara entre ellos.
    En particular, el Silhouette Score es un valor alto, lo que indica que los datos están bien agrupados y que cada punto de datos está cerca de los puntos de datos de su propio cluster y lejos de los puntos de datos de los otros clusters.
    El Calinski-Harabasz Score también es un valor alto, lo que indica que los clusters son compactos y que están bien separados entre sí.
    El Davies-Bouldin Index es un valor bajo, lo que indica que los clusters son de tamaño similar y que están bien distribuidos.
    En conclusión, el modelo KMeans es un modelo muy efectivo para el rendimiento de jugadores de fútbol. Este modelo es capaz de agrupar los datos en clusters de manera eficiente y clara, lo que permite identificar patrones y tendencias relevantes.
