```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = '../DataSets/2022-2023FootballPlayerStats.csv'
df = pd.read_csv(ruta_archivo, encoding='Windows-1252', sep=';')

# Ver las primeras filas del dataset
print(df.head())
```

```python
# Obtener la cantidad de filas y columnas del DataFrame
num_filas = df.shape[0]
num_columnas = df.shape[1]

print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)
```

```python
# Obtener información sobre el conjunto de datos
print(df.info())

# Calcular estadísticas descriptivas
print(df.describe())
```

```python
# Visualización univariada - Histograma de edad
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribución de la Edad de los Jugadores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Resumen numérico básico de la edad
edad_resumen = df['Age'].describe()
print("Resumen de la Edad:")
print(edad_resumen)
```

```python
# Resumen numérico básico: Estadísticas de goles y asistencias
goles_stats = df['Goals'].describe()
asistencias_stats = df['Assists'].describe()

print('Estadísticas de Goles:')
print(goles_stats)
print('\nEstadísticas de Asistencias:')
print(asistencias_stats)
```

```python
# Visualización bivariada: Relación entre posición y asistencias
# Ordenar el DataFrame por la columna 'Assists' de forma descendente
df_sorted = df.sort_values('Assists', ascending=False)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sorted, x='Pos', y='Assists')
plt.title('Relación entre posición y Asistencias (ordenado de mayor a menor)')
plt.xlabel('Posición')
plt.ylabel('Asistencias')
plt.show()
```

```python
# Punto fuera de rango 
# Obtener el índice del máximo valor en la columna 'Assists'
max_assist_index = df['Assists'].idxmax()

# Obtener el nombre del jugador máximo asistidor
max_assist_player = df.loc[max_assist_index, 'Player']

# Obtener el nombre del equipo máximo asistidor
max_assist_squad = df.loc[max_assist_index, 'Squad']

# Obtener la cantidad de goles del jugador máximo goleador
max_assists = df.loc[max_assist_index, 'Assists']

print("El jugador máximo asistidor es:", max_assist_player)
print("Cantidad de asistencias:", max_assists)
print("Juega en el equipo:", max_assist_squad)
```

```python
#Resultados Visualización bivariada
'''
En el grafico anterior se observa una mayor de cantidad de asistencias por parte de la posición MF seguido por la posición FW.

'''
```

```python
# Visualización trivariada: Relación entre edad, goles y posición
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Age', y='Goals', hue='Pos')
plt.title('Relación entre Edad, Goles y Posición')
plt.xlabel('Edad')
plt.ylabel('Goles')
plt.legend(title='Posición')
plt.show()
```

```python
#Resultados Visualización trivariada
'''
En el grafico anterior no se observa correlación entre edad y cantidad de goles. Si un rango minimo (20 años) y maximo (36 años)
de promedio de goles.
Se observa ademas una correlación entre posición y cantidad de goles, la posición FW es la que proporciona mayor cantidad 
de goles,lo cual es logico ya que es la posición mas cercana al arco contrario.
'''
```

```python
# Punto fuera de rango 
# Obtener el índice del máximo valor en la columna 'Goles'
max_goal_index = df['Goals'].idxmax()

# Obtener el nombre del jugador máximo goleador
max_goal_player = df.loc[max_goal_index, 'Player']

# Obtener la cantidad de goles del jugador máximo goleador
max_goles = df.loc[max_goal_index, 'Goals']

print("El jugador máximo goleador es:", max_goal_player)
print("Cantidad de goles:", max_goles)
```
