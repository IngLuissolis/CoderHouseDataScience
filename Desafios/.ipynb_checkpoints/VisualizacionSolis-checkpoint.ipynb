```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = '../DataSets/2022-2023FootballPlayerStats.csv'
df = pd.read_csv(ruta_archivo, encoding='Windows-1252', sep=';')

# Ver las primeras filas del dataset
print(df.head())
```

```python
# Obtener la cantidad de filas y columnas del DataFrame
num_filas = df.shape[0]
num_columnas = df.shape[1]

print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)
```

```python
#Matplotlib - Grafico 1 - Gráfico de barras de la cantidad de jugadores por posición:
df["Pos"].value_counts().plot(kind="bar")
plt.title("Cantidad de jugadores por posición")
plt.xlabel("Posición")
plt.ylabel("Cantidad de jugadores")
plt.show()
```

```python
#Matplotlib - Grafico 1 - Gráfico de barras del número de jugadores por país:
top_10_nations = df["Nation"].value_counts().head(10)
plt.bar(top_10_nations.index, top_10_nations.values)
plt.title("Número de jugadores por país")
plt.xlabel("País")
plt.ylabel("Número de jugadores")
plt.show()
```

```python
#Matplotlib - Grafico 3 - Gráfico de dispersión de la relación entre edad y minutos jugados:
plt.scatter(df["Age"], df["Min"])
plt.title("Relación entre edad y minutos jugados")
plt.xlabel("Edad")
plt.ylabel("Minutos jugados")
plt.show()
```

```python
#SeaBorn - Grafico 1 - Gráfico de barras del número de jugadores por posición:
sns.countplot(x="Pos", data=df)
plt.title("Número de jugadores por posición")
plt.xlabel("Posición")
plt.ylabel("Número de jugadores")
plt.show()
```

```python
#SeaBorn - Grafico 2 - Gráfico de densidad de la distribución de la edad de los jugadores:
sns.kdeplot(df["Age"], shade=True)
plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.show()
```

```python
#SeaBorn - Grafico 3 - Gráfico de caja de la relación entre posición y minutos jugados:
sns.boxplot(x="Pos", y="Min", data=df)
plt.title("Relación entre posición y minutos jugados")
plt.xlabel("Posición")
plt.ylabel("Minutos jugados")
plt.show()
```
