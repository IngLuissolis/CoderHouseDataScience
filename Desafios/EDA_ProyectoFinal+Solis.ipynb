## Datos de Interes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = '../DataSets/european_leagues_player_stats/Big5-statsStats22-23_Corr.csv'

df = pd.read_csv(ruta_archivo, delimiter=',')

# Ver las primeras filas del dataset
df.head()
```

## Analisis Exploratorio de Datos (EDA)

```python
# Obtener informaciÃ³n sobre el conjunto de datos
df.info()

# Calcular estadÃ­sticas descriptivas
df.describe()
```

```python
# VisualizaciÃ³n univariada - Histograma de edad
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', kde=True)
plt.title('DistribuciÃ³n de la Edad de los Jugadores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# VisualizaciÃ³n bivariada: RelaciÃ³n entre posiciÃ³n y asistencias
# Ordenar el DataFrame por la columna 'Assists' de forma descendente
df_sorted = df.sort_values('Ast', ascending=False)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sorted, x='Pos', y='Ast')
plt.title('RelaciÃ³n entre posiciÃ³n y Asistencias (ordenado de mayor a menor)')
plt.xlabel('PosiciÃ³n')
plt.ylabel('Asistencias')
plt.show()
```

### Resultados VisualizaciÃ³n bivariada

    En el grafico anterior se observa una mayor de cantidad de asistencias por parte de la posiciÃ³n MF 
    seguido por la posiciÃ³n FW.

```python
# Punto fuera de rango 
# Obtener el Ã­ndice del mÃ¡ximo valor en la columna 'Assists'
max_assist_index = df['Ast'].idxmax()

# Obtener el nombre del jugador mÃ¡ximo asistidor
max_assist_player = df.loc[max_assist_index, 'Player']

# Obtener el nombre del equipo mÃ¡ximo asistidor
max_assist_squad = df.loc[max_assist_index, 'Squad']

# Obtener la cantidad de goles del jugador mÃ¡ximo goleador
max_assists = df.loc[max_assist_index, 'Ast']

print("El jugador mÃ¡ximo asistidor es:", max_assist_player)
print("Cantidad de asistencias:", max_assists)
print("Juega en el equipo:", max_assist_squad)
```

```python
# VisualizaciÃ³n trivariada: RelaciÃ³n entre edad, goles y posiciÃ³n
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Age', y='Gls', hue='Pos')
plt.title('RelaciÃ³n entre Edad, Goles y PosiciÃ³n')
plt.xlabel('Edad')
plt.ylabel('Goles')
plt.legend(title='PosiciÃ³n')
plt.show()
```

### Resultados VisualizaciÃ³n trivariada

En el grafico anterior no se observa correlaciÃ³n entre edad y cantidad de goles. Si un rango minimo (20 aÃ±os) y maximo (36 aÃ±os) de promedio de goles.

Se observa ademas una correlaciÃ³n entre posiciÃ³n y cantidad de goles, la posiciÃ³n FW es la que proporciona mayor cantidad 
de goles,lo cual es logico ya que es la posiciÃ³n mas cercana al arco contrario.

# 1. Â¿CuÃ¡les son los jugadores mÃ¡s efectivos en tÃ©rminos de goles marcados? Â¿Existen diferencias significativas entre los distintos equipos en cuanto a la producciÃ³n de goles?

```python
# Agrupar los datos por jugador y sumar los goles marcados por cada jugador
jugadores_goles = df.groupby(['Player', 'Squad'])['Gls'].sum().reset_index()

# Ordenar los jugadores por la cantidad de goles en orden descendente
jugadores_top15 = jugadores_goles.nlargest(15, 'Gls')

# Mostrar los 15 jugadores mÃ¡s efectivos en tÃ©rminos de goles marcados
jugadores_top15.head(15)
```

```python
# Agrupar los datos por jugador y sumar las asistencias por cada jugador
jugadores_goles = df.groupby(['Unnamed: 0_level_0_Player', 'Team'])['Performance_Ast'].sum().reset_index()

# Ordenar los jugadores por la cantidad de asistencias en orden descendente
jugadores_top15 = jugadores_goles.nlargest(15, 'Performance_Ast')

# Mostrar los 15 jugadores mÃ¡s efectivos en tÃ©rminos de asistencias
jugadores_top15.head(15)
```

```python
# Agrupar los datos por equipo y sumar los goles marcados por cada equipo
equipos_goles = df.groupby('Team')['Performance_Gls'].sum().reset_index()

# Ordenar los equipos por la cantidad de goles en orden descendente
equipos_top10 = equipos_goles.nlargest(10, 'Performance_Gls')

# Mostrar los diez equipos con mayor cantidad de goles
print(equipos_top10)

# GrÃ¡fico de barras para visualizar la producciÃ³n de goles de los equipos top 10
plt.figure(figsize=(10, 6))
plt.bar(equipos_top10['Team'], equipos_top10['Performance_Gls'])
plt.xlabel('Equipo')
plt.ylabel('Goles')
plt.title('ProducciÃ³n de goles de los diez equipos con mayor cantidad de goles')
plt.xticks(rotation=90)
plt.show()
```

```python
# Filtrar los datos para obtener solo los jugadores del equipo Bayern Munich
bayern_munich = df[df['Team'] == 'Bayern Munich']

# Agrupar los datos por jugador y sumar los goles marcados por cada jugador
bayer_munich_goles = bayern_munich.groupby(['Unnamed: 0_level_0_Player', 'Unnamed: 2_level_0_Pos', 'Progression_PrgC', 'Progression_PrgP', 'Progression_PrgR'])['Performance_Gls'].sum().reset_index()
bayer_munich_goles = bayer_munich_goles[bayer_munich_goles['Performance_Gls'] >= 1]

# Ordenar los jugadores por la cantidad de goles en orden descendente
bayer_munich_goles = bayer_munich_goles.sort_values(by='Performance_Gls', ascending=False)

# Mostrar los jugadores del equipo Bayer Munich con mÃ¡s de 1 gol
print("Jugadores del equipo Bayer Munich con al menos 1 gol:")
print(bayer_munich_goles)

# Calcular el total de goles del equipo Bayer Munich
bayer_munich_total_goles = bayer_munich_goles['Performance_Gls'].sum()
print("Total de goles del equipo Bayer Munich:", bayer_munich_total_goles)

print("")
```
