### Variables Sinteticas

    Sh_per_90: Número de tiros por 90 minutos jugados. Esta variable mide la frecuencia con la que un jugador dispara al arco. Se puede calcular dividiendo la variable Sh por la variable 90s.
    
    SoT_per_90: Número de oportunidades de gol por 90 minutos jugados. Esta variable mide la frecuencia con la que un jugador crea oportunidades de gol. Se puede calcular dividiendo la variable SoT por la variable 90s.
    
    G_per_SoT: Número de goles por oportunidad de gol. Esta variable mide la efectividad de un jugador a la hora de convertir oportunidades de gol. Se puede calcular dividiendo la variable Gls por la variable SoT.


```python
import pandas as pd
import numpy as np

ruta_archivo = '../DataSets/european_leagues_player_stats/combined_data.csv'

# Read the combined data into a Pandas DataFrame
df = pd.read_csv(ruta_archivo, delimiter=',')

df = df[['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'Starts', '90s', 'Gls', 'Sh', 'SoT', 'FK', 'Dist']]
```

```python
# Create the `SoT_per_90` variable
df['Sh_per_90'] = df['Sh'] / df['90s']

# Create the `SoT_per_90` variable
df['SoT_per_90'] = df['SoT'] / df['90s']

# Create the `G_per_SoT` variable
df['G_per_SoT'] = df['Gls'] / df['SoT']

# Eliminar valores NaN
df['SoT_per_90'].replace(np.NAN, 0, inplace=True)
df['G_per_SoT'].replace(np.NAN, 0, inplace=True)
df['Sh_per_90'].replace(np.NAN, 0, inplace=True)

# Eliminar valores infinitos
df['SoT_per_90'].replace(np.inf, 0, inplace=True)
df['G_per_SoT'].replace(np.inf, 0, inplace=True)
df['Sh_per_90'].replace(np.inf, 0, inplace=True)

# Eliminar valores demasiado grandes
df['SoT_per_90'].replace(np.nan, 0, inplace=True)
df['G_per_SoT'].replace(np.nan, 0, inplace=True)
df['Sh_per_90'].replace(np.nan, 0, inplace=True)

# Save the DataFrame to a CSV file
df.to_csv('../DataSets/european_leagues_player_stats/sintetic_data.csv', index=False)

ruta_archivo = '../DataSets/european_leagues_player_stats/sintetic_data.csv'

# Read the combined data into a Pandas DataFrame
df = pd.read_csv(ruta_archivo, delimiter=',')

# Get information about the DataFrame
df.info()

# Calculate descriptive statistics
df.describe()

# See the first few rows of the DataFrame
df.head()
```
