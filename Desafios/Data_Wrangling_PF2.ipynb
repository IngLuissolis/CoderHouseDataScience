```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ruta_archivo = '../DataSets/european_leagues_player_stats/Big5-statsStats22-23.csv'

df = pd.read_csv(ruta_archivo, delimiter=',')

# Obtener información sobre el conjunto de datos
df.info()

# Calcular estadísticas descriptivas
df.describe()

# Ver las primeras filas del dataset
df.head()
```

#### Borrar Filas
        
        Eliminar las filas con Nation = NaN
        
        Eliminar las filas con xG = NaN


```python
# Cambiar el valor del player David Ozoh y columna "Nation"
df.loc[2010, 'Nation'] = 'en ENG'

# Mostrar solo la fila 1875 del DataFrame
fila_2010 = df.iloc[2010:2013]
print(fila_2010)
```

```python
# Eliminar las filas con variable Nation = NaN
df = df.dropna(subset=['Nation'])

# Eliminar las filas con variable xG = NaN
df = df.dropna(subset=['xG'])

# Obtener información sobre el conjunto de datos
df.info()

# Calcular estadísticas descriptivas
df.describe()

# Ver las primeras filas del dataset
df.head()
```

## Guardar los datos en un archivo nuevo CSV
