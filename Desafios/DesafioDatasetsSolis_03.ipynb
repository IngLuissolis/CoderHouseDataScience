```python
import pandas as pd
import matplotlib.pyplot as plt

ruta_archivo = '../DataSets/2022-2023Top5FootballLeaguesPlayerStats.csv'
df = pd.read_csv(ruta_archivo, encoding='Windows-1252')

# Ver las primeras filas del dataset
print(df.head())
```

```python
# Obtener la cantidad de filas y columnas del DataFrame
num_filas = df.shape[0]
num_columnas = df.shape[1]

print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)
```

```python
# contar la cantidad de jugadores por nacionalidad (10 nacionalidades con mayor cantidad de jugadores)
nacionalidades = df['Nation'].value_counts()

# graficar el histograma
nacionalidades.head(10).plot.bar()
plt.title('Top 10 Nacionalidades con Mayor Cantidad de Jugadores')
plt.xlabel('Nacionalidad')
plt.ylabel('Cantidad de Jugadores')
plt.show()
```

```python
# Seleccionar columnas relevantes y filtrar por Nation ARG
cols = ["Player name ", "Nation", "Squad", "Position", "Goal", "Assist"]
df_ARG = df.loc[df["Nation"] == "ARG", cols]

# Ordenar por cantidad de goles en orden descendente
df_ARG = df_ARG.sort_values(by=["Goal"], ascending=False)

# Tomar los primeros 10 jugadores
top_10_ARG = df_ARG.head(10)

# Mostrar tabla
print(top_10_ARG.to_string(index=False))
```
