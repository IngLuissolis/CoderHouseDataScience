# 1. ¿Cuáles son los jugadores más efectivos en términos de goles marcados? ¿Existen diferencias significativas entre los distintos equipos en cuanto a la producción de goles?

```python
# Agrupar los datos por jugador y sumar los goles marcados por cada jugador
jugadores_goles = df.groupby(['Player', 'Squad'])['Gls'].sum().reset_index()

# Ordenar los jugadores por la cantidad de goles en orden descendente
jugadores_top15 = jugadores_goles.nlargest(15, 'Gls')

# Mostrar los quince jugadores más efectivos en términos de goles marcados
jugadores_top15.head(20)
```

```python
# Agrupar los datos por equipo y sumar los goles marcados por cada equipo
equipos_goles = df.groupby('Squad')['Gls'].sum().reset_index()

# Ordenar los equipos por la cantidad de goles en orden descendente
equipos_top10 = equipos_goles.nlargest(10, 'Gls')

# Mostrar los diez equipos con mayor cantidad de goles
print(equipos_top10)

# Gráfico de barras para visualizar la producción de goles de los equipos top 10
plt.figure(figsize=(10, 6))
plt.bar(equipos_top10['Squad'], equipos_top10['Gls'])
plt.xlabel('Equipo')
plt.ylabel('Goles')
plt.title('Producción de goles de los diez equipos con mayor cantidad de goles')
plt.xticks(rotation=90)
plt.show()
```

```python
# Filtrar los datos para obtener solo los jugadores de los equipos deseados
teams = ['Bayern Munich', 'Paris S-G', 'Manchester City']
filtered_df = df[df['Squad'].isin(teams)]

# Crear un gráfico de boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_df, x='Squad', y='Gls')
plt.title('Distribución de Goles por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Goles')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.show()
```

```python
# Filtrar los datos para obtener solo los jugadores de los equipos deseados
teams = ['Bayern Munich', 'Paris S-G', 'Manchester City']
filtered_df = df[df['Squad'].isin(teams)]

# Agrupar los datos por equipo y jugador, sumar los goles y ordenar de manera descendente
top_scorers = filtered_df.groupby(['Squad', 'Player'])['Gls'].sum().reset_index()
top_scorers = top_scorers.sort_values(by=['Squad', 'Gls'], ascending=[True, False])

# Obtener a los 5 máximos goleadores por equipo
top_bayern_munich = top_scorers[top_scorers['Squad'] == 'Bayern Munich'].head(5)
top_paris_sg = top_scorers[top_scorers['Squad'] == 'Paris S-G'].head(5)
top_manchester_city = top_scorers[top_scorers['Squad'] == 'Manchester City'].head(5)

# Mostrar a los 5 goleadores máximos por equipo
print("Los 5 goleadores máximos del Bayern Munich:")
print(top_bayern_munich)
print("\nLos 5 goleadores máximos del Paris S-G:")
print(top_paris_sg)
print("\nLos 5 goleadores máximos del Manchester City:")
print(top_manchester_city)
```

### Conclusiones

El rendimiento goleador de los equipos Manchester City, Paris S-G y Bayer Munich varía en función de la dependencia de ciertos jugadores. Mientras que el Manchester City se apoya fuertemente en Erling Haaland para la anotación de goles, Paris S-G confía en Kylian Mbappé, Neymar y Lionel Messi. Por otro lado, el equipo Bayer Munich muestra una distribución más equitativa en cuanto a la producción de goles, con varios jugadores contribuyendo de manera similar en términos de goles anotados".

Esta hipótesis sugiere que la dependencia de un solo jugador para la producción de goles puede influir en el rendimiento goleador de un equipo. Mientras que el Manchester City depende en gran medida de Erling Haaland, Paris S-G cuenta con varios jugadores clave, y el Bayer Munich muestra una mayor distribución en la producción de goles entre sus jugadores.

Si uno fuera rival de Manchester City, el planteo seria realizar una fuerte marcación contra Erling Haaland evitando que le llegue el balon.

# 2. ¿Existe una relación entre la precisión de los pases de un jugador y su capacidad para crear oportunidades de gol para su equipo?

#### PrgR: Pases progresivos por partido

```python
# Agregar la columna 'PrgR' a los DataFrames de los máximos goleadores
top_bayern_munich = top_bayern_munich.merge(filtered_df[['Player', 'PrgR']], on='Player', how='left')
top_paris_sg = top_paris_sg.merge(filtered_df[['Player', 'PrgR']], on='Player', how='left')
top_manchester_city = top_manchester_city.merge(filtered_df[['Player', 'PrgR']], on='Player', how='left')

# Mostrar los 3 goleadores máximos por equipo con la columna 'PrgR'
print("Los 5 goleadores máximos del Bayern Munich:")
print(top_bayern_munich)
print("\nLos 5 goleadores máximos del Paris S-G:")
print(top_paris_sg)
print("\nLos 5 goleadores máximos del Manchester City:")
print(top_manchester_city)
```

```python
# Crear un gráfico de dispersión para Bayern Munich
plt.figure(figsize=(10, 6))
plt.scatter(top_bayern_munich['PrgR'], top_bayern_munich['Gls'], label='Bayern Munich', color='red')
for i, player in enumerate(top_bayern_munich['Player']):
    plt.annotate(player, (top_bayern_munich['PrgR'].iloc[i], top_bayern_munich['Gls'].iloc[i]), fontsize=8)

# Crear un gráfico de dispersión para Paris S-G
plt.scatter(top_paris_sg['PrgR'], top_paris_sg['Gls'], label='Paris S-G', color='blue')
for i, player in enumerate(top_paris_sg['Player']):
    plt.annotate(player, (top_paris_sg['PrgR'].iloc[i], top_paris_sg['Gls'].iloc[i]), fontsize=8)

# Crear un gráfico de dispersión para Manchester City
plt.scatter(top_manchester_city['PrgR'], top_manchester_city['Gls'], label='Manchester City', color='green')
for i, player in enumerate(top_manchester_city['Player']):
    plt.annotate(player, (top_manchester_city['PrgR'].iloc[i], top_manchester_city['Gls'].iloc[i]), fontsize=8)

# Configurar el gráfico
plt.title('Gráfico de Dispersión Pases progresivos por partido vs. Goles')
plt.xlabel('Pases progresivos por partido')
plt.ylabel('Goles')
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Calcular el coeficiente de correlación entre 'PrgR' y 'Gls' para todos los jugadores
correlation = np.corrcoef(df['PrgR'], df['Gls'])[0, 1]

print(f"El coeficiente de correlación entre la precisión de los pases (PrgR) y la capacidad para crear oportunidades de gol (Gls) es: {correlation:.2f}")
```

## Conclusiones

    el coeficiente de correlación de 0.65 sugiere una correlación positiva moderada entre la precisión de los pases y la capacidad para crear oportunidades de gol. ## Creación de Variables sinteticas

```python
# Create the `SoT_per_90` variable
df['Sh_per_90'] = df['Sh'] / df['90s']

# Create the `SoT_per_90` variable
df['SoT_per_90'] = df['SoT'] / df['90s']

# Create the `G_per_SoT` variable
df['G_per_SoT'] = df['Gls'] / df['SoT']

# Eliminar valores NaN
df['SoT_per_90'].replace(np.NAN, 0, inplace=True)
df['G_per_SoT'].replace(np.NAN, 0, inplace=True)
df['Sh_per_90'].replace(np.NAN, 0, inplace=True)

# Eliminar valores infinitos
df['SoT_per_90'].replace(np.inf, 0, inplace=True)
df['G_per_SoT'].replace(np.inf, 0, inplace=True)
df['Sh_per_90'].replace(np.inf, 0, inplace=True)

# Eliminar valores demasiado grandes
df['SoT_per_90'].replace(np.nan, 0, inplace=True)
df['G_per_SoT'].replace(np.nan, 0, inplace=True)
df['Sh_per_90'].replace(np.nan, 0, inplace=True)

# Save the DataFrame to a CSV file
df.to_csv('../DataSets/european_leagues_player_stats/sintetic_data.csv', index=False)

ruta_archivo = '../DataSets/european_leagues_player_stats/sintetic_data.csv'

# Read the combined data into a Pandas DataFrame
df = pd.read_csv(ruta_archivo, delimiter=',')

# Get information about the DataFrame
df.info()

# Calculate descriptive statistics
df.describe()

# See the first few rows of the DataFrame
df.head()
```

## Metodo K-Fold Cross Validation para Modelo Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# Select the variables to use
features = ['Starts', 'Sh', 'SoT', 'FK', 'Dist', 'Sh_per_90', 'SoT_per_90', 'G_per_SoT']

# Standardize the data
df_scaled = (df[features] - df[features].mean()) / df[features].std()

# Define the KFold object - K = 10
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

# Create a list to store the RMSEs
rmses = []

# Iterate over the KFold splits
for train_index, test_index in kfold.split(df_scaled):
    # Split the data into training and test sets
    X_train, X_test = df_scaled.iloc[train_index], df_scaled.iloc[test_index]
    y_train, y_test = df['Gls'].iloc[train_index], df['Gls'].iloc[test_index]

    # Fit the random forest regressor model
    rf = RandomForestRegressor(n_estimators=100)
    rf.fit(X_train, y_train)

    # Evaluate the model on the test set
    y_pred = rf.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    rmses.append(rmse)

# Print the average RMSE
print("Average RMSE:", np.mean(rmses))
```

### Las conclusiones para el método K-Fold Cross Validation son las siguientes:

    El RMSE obtenido es mayor que el obtenido con el modelo original, lo que sugiere que el modelo es menos preciso cuando se utiliza K-Fold Cross Validation.

    Este resultado puede deberse a que K-Fold Cross Validation utiliza una validación cruzada estratificada, lo que significa que cada partición tiene una distribución de clases similar a la distribución de clases en el conjunto de datos completo. En este caso, el conjunto de datos tiene dos clases: jugadores que marcaron goles y jugadores que no marcaron goles. Al utilizar K-Fold Cross Validation, nos aseguramos de que cada partición tenga un número similar de jugadores que marcaron goles y jugadores que no marcaron goles. Esto puede ayudar a mejorar la precisión del modelo de aprendizaje automático, pero también puede introducir sesgos en los resultados.
    
    
### Para mejorar la precisión del modelo con K-Fold Cross Validation, podemos considerar las siguientes soluciones:

    Utilizar un número mayor de particiones. Esto ayudará a reducir el sesgo introducido por la validación cruzada estratificada.
    
    Utilizar un método de validación no estratificado. Esto puede ayudar a mejorar la precisión del modelo en general, pero también puede introducir sesgos en los resultados.
    
    Utilizar un modelo de aprendizaje automático diferente. Algunos modelos, como los modelos de regresión lineal, pueden ser más precisos que los modelos de regresión de bosque aleatorio cuando se utilizan métodos de validación cruzada.


## Metodo Stratified K-Fold para Modelo Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error

# Select the variables to use
features = ['Starts', 'Sh', 'SoT', 'FK', 'Dist', 'Sh_per_90', 'SoT_per_90', 'G_per_SoT']

# Standardize the data
df_scaled = (df[features] - df[features].mean()) / df[features].std()

# Define the StratifiedKFold object - K = 10
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Create a list to store the RMSEs
rmses = []

# Iterate over the StratifiedKFold splits
for train_index, test_index in skf.split(df_scaled, df['Gls']):
    # Split the data into training and test sets
    X_train, X_test = df_scaled.iloc[train_index], df_scaled.iloc[test_index]
    y_train, y_test = df['Gls'].iloc[train_index], df['Gls'].iloc[test_index]

    # Fit the random forest regressor model
    rf = RandomForestRegressor(n_estimators=100)
    rf.fit(X_train, y_train)

    # Evaluate the model on the test set
    y_pred = rf.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    rmses.append(rmse)

# Print the average RMSE
print("Average RMSE:", np.mean(rmses))
```

### Las conclusiones para el método Stratified K-Fold son las siguientes:

    El RMSE obtenido es mayor que el obtenido con el modelo Random Forest Regressor, lo que sugiere que el modelo es menos preciso.
    
    Este resultado es esperado, ya que Stratified K-Fold toma en cuenta la distribución de las clases en los datos. En este caso, el conjunto de datos tiene dos clases: jugadores que marcaron goles y jugadores que no marcaron goles. 
    
    Al utilizar Stratified K-Fold, nos aseguramos de que cada partición tenga un número similar de jugadores que marcaron goles y jugadores que no marcaron goles. Esto puede ayudar a mejorar la precisión del modelo de aprendizaje automático, pero también puede reducir la precisión en los casos en que el modelo está bien entrenado para predecir una clase en particular.
    
### Para mejorar la precisión del modelo utilizando Stratified K-Fold:

    podemos aumentar el número de particiones. Esto ayudará a que el modelo tenga más datos de entrenamiento para cada clase.

    Otra opción es utilizar un método de validación cruzada que no tome en cuenta la distribución de las clases. Por ejemplo, podemos utilizar K-Fold con un número de particiones impar. Esto ayudará a que el modelo tenga una partición completa para cada clase.
